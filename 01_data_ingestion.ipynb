# 01 Data Ingestion

Load the 10 provided Bluestock datasets, inspect shape/dtypes/head, and validate AMFI scheme-code coverage.

```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "data").exists() and (PROJECT_ROOT.parent / "data").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
RAW_DIR = PROJECT_ROOT / "data" / "raw"
PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
REPORTS_DIR = PROJECT_ROOT / "reports"
CHARTS_DIR = REPORTS_DIR / "charts"
sns.set_theme(style="whitegrid")
PROJECT_ROOT
```

```python
raw_files = sorted(RAW_DIR.glob("*.csv"))
datasets = {p.stem: pd.read_csv(p) for p in raw_files if not p.name.startswith("live_nav")}
[(name, df.shape) for name, df in datasets.items()]
```

```python
for name, df in datasets.items():
    print(f"\n{name}: {df.shape}")
    display(df.head())
    display(df.dtypes.rename("dtype").to_frame())
```

```python
fund = datasets["01_fund_master"]
nav = datasets["02_nav_history"]
missing_in_nav = sorted(set(fund["amfi_code"]) - set(nav["amfi_code"]))
extra_in_nav = sorted(set(nav["amfi_code"]) - set(fund["amfi_code"]))
print("Fund master schemes:", fund["amfi_code"].nunique())
print("NAV schemes:", nav["amfi_code"].nunique())
print("Missing in NAV:", missing_in_nav)
print("Extra in NAV:", extra_in_nav)
print(nav.groupby("amfi_code").size().describe())
```
